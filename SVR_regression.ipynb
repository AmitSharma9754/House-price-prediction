### Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Read data

```python
data = pd.read_csv('house_price_prediction.csv')
```

### Handle NULL values

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
data = imputer.fit_transform(data)
```

```python
X = data[:, :-1]
y = data[:, -1]
```

```python
y = y.reshape(len(y), 1)
```

### Split dataset into train and test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 0)
```

### Apply Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
X_sc = StandardScaler()
y_sc = StandardScaler()
X_train = X_sc.fit_transform(X_train)
X_test = X_sc.fit_transform(X_test)
y_train = y_sc.fit_transform(y_train)
y_test = y_sc.fit_transform(y_test)
```

### Train our model

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X_train, y_train)
```

### Test our model

```python
y_pred = y_sc.inverse_transform(regressor.predict(X_test).reshape(-1,1))
```

### Evaluate Model on R2

```python
from sklearn.metrics import r2_score
r2_score(y_sc.inverse_transform(y_test), y_pred) 
```
